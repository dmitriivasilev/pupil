```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pupil_area = np.load('pupil_area.npy')
pupil_center = np.load('pupil_center.npy')
```

```python
plt.figure(figsize=(20,10))
plt.plot(pupil_area)
plt.plot([x*10 for x,y in pupil_center])
plt.plot([y*10 for x,y in pupil_center])
```

```python
plt.figure(figsize=(20,10))
plt.plot(pupil_area)
plt.figure(figsize=(20,10))
plt.plot(np.diff(pupil_area))
```

```python
plt.hist(np.diff([x for x in pupil_area if abs(x)<5000]))
```

```python
pupil_area_diff = np.diff(pupil_area)
```

```python
len(pupil_area)
```

```python
len(pupil_area_diff)
```

```python
len([x for x in pupil_area_diff if abs(x)>500])/len(pupil_area_diff)
```

```python
filtered_pupil_area = [x if abs(pupil_area_diff[i])<500 else 0 for i,x in enumerate(pupil_area[:-1])]
```

```python
len(filtered_pupil_area)
```

```python
plt.figure(figsize=(20,10))
plt.plot(filtered_pupil_area)
```

```python
restored_pupil_area = [x if x>0 else np.mean(filtered_pupil_area[i-5:i+5]) for i,x in enumerate(filtered_pupil_area)]
```

```python
plt.figure(figsize=(20,10))
plt.plot(restored_pupil_area)
```
