```python
#we filter them by length 
# the line below is a list comprehention (google it)
#basicaly it creates another list taking elements from contours which have length < 1000 and > 200
filtered_contours = [cnt for cnt in contours if (len(cnt)<1000) and (len(cnt)>200)]
```

```python
#lets draw them
filtered_contour_image = cv2.drawContours(raw_image, filtered_contours, -1, (255,255,255), 3)

plt.imshow(filtered_contour_image)
```

```python
#we see that all the contoures are quite similar in size but differ in shape 
# the plan is to find one which is the most similar to ellipse
```

```python
ellipses = [] # we create an empty list to fill later
for cnt in filtered_contours: # for each contour from out list
    try: # we try to apply ellipse fitting
        ellipses.append(cv2.fitEllipse(cnt)) #and add the result to list of ellipses
    except: # if not
        pass # whatever
#P.S.: remember, I didn't teach you to do that
```

```python
# it means that at least our thing didn't go completely wrong
```

```python
for ellipse in ellipses:
    plt.figure()
    plt.imshow(cv2.ellipse(raw_image.copy(),ellipse,(255,255,255),2))
```

```python
# we want to get the one which is the most similar to ellipse
# we introduse "loss" which is the difference in the are of fitted ellipse and contour 
# so if this loss id close to zero then the shape of contour is close to an ellipse
# which is completely wrong but it will work for now

loss = [] #emply list to fill the loss values
for ellipse,contour in zip(ellipses,filtered_contours): # for each ellipse and contours from our ellipses and contours 
                                                        # (google zip function)
    loss.append(abs((np.pi/4 * ellipse[1][0] * ellipse[1][1])-(cv2.contourArea(contour))))
#difference between ellipse area ^              and contour area        ^                       
```

```python
loss # we have three values
```

```python
#we draw argmin of loss (argmin returns the index of minimum values of array/list)
plt.imshow(cv2.ellipse(raw_image,ellipses[np.argmin(loss)],(255,255,255),2))
```

```python
def find_pupple(filename):
    raw_image = cv2.imread(filename,0)
    
    ret,thresh = cv2.threshold(raw_image,50,255,0)
    image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    contour_image = cv2.drawContours(raw_image.copy(), contours, -1, (255,255,255), 3)

    filtered_contours = [cnt for cnt in contours if (len(cnt)<1000) and (len(cnt)>200)]

    ellipses = [] 
    for cnt in filtered_contours: 
        try: 
            ellipses.append(cv2.fitEllipse(cnt)) 
        except: 
            pass 

        loss = [] 

    for ellipse,contour in zip(ellipses,filtered_contours): 
        loss.append(abs((np.pi/4 * ellipse[1][0] * ellipse[1][1])-(cv2.contourArea(contour))))
    

    return cv2.ellipse(raw_image,ellipses[np.argmin(loss)],(255,255,255),2)
```
